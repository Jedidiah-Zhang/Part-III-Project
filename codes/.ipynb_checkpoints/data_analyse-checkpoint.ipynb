## ============= Data Preparation =============

### 1. Characteristics of Links
        | ID | Length | Width | Class |
    
**ID (string):** Unique ID for a specific slice (links) of road.

**Length (int):** The length of the link.

**Width (int):** The width of the link.

**Class (int):** The road class (all 1 in this dataset hence not in use).


```python
links = {}
with open("..\datasets\gy_link_info.txt") as csvfile:
    csv_reader = csv.reader(csvfile, delimiter=";")
    header = next(csv_reader)
    for row in csv_reader:
        ID = row[0]
        length = int(row[1])
        width = int(row[2])
        links[ID] = [length, width]
```

### 2. Map (topology of links)
        | ID | In-links | Out-links |

**In-links:** list of upstream link IDs

**Out-links:** list of downstream link IDs

```python
data = []
with open("..\datasets\gy_link_top.txt") as csvfile:
    csv_reader = csv.reader(csvfile, delimiter=";")
    header = next(csv_reader)
    for row in csv_reader:
        ID = row[0]
        in_links = [-1 if link == "" else int(link) for link in row[1].split("#")]
        out_links = [0 if link == "" else int(link) for link in row[2].split("#")]
        data.append([ID, in_links, out_links])
MAP = dict(zip([ID[0] for ID in data], [X[1:] for X in data]))
```

### 3. Travel Time Data
        | ID | Date | Time Interval | Travel Time |

**Date:** Example: 2016-05-21

**Time Interval:** Example: [2016-05-21 17:50:00,2016-05-21 17:52:00)

**Travel Time (s):** The average time in seconds of all vehicles entered this link with in the time interval.

```python
def load_data(filename):
    data = []
    with open(filename) as csvfile:
        csv_reader = csv.reader(csvfile, delimiter=";")
        next(csvfile)
        for row in csv_reader:
            ID = row[0]
            date = [int(x) for x in row[1].split("-")]
            [start, end] = row[2].split(",")
            time_interval = [int(i) for i in start[12:].split(":")]
            travel_time = float(row[3])
            data.append([ID, date[0], date[1], date[2], time_interval[0] * 60 + time_interval[1] + 1, travel_time])
    return data
```

```python
data = load_data("..\datasets\gy_link_travel_time_part1.txt")
```

```python
sorted_data = {}
for record in data:
    if record[0] in sorted_data.keys():
        sorted_data[record[0]].append(record[1:])
    else: sorted_data[record[0]] = [record[1:]]
```

```python
lengths = []
widths = []
avrs = []
var_s = []

for link in sorted_data.keys():
    lengths.append(links[link][0])
    widths.append(links[link][1])
    records = np.array(sorted_data[link])
    avrs.append(np.sum(records[:, 4]) / len(records))
    var_s.append(np.var(records[:, 4]))

```

```python
def avr_time_of_day(data, link):
    avr_times = {}
    for i in range(1, 1441, 2):
        sum = count = 0
        for each in data[link]:
            if each[3] == i:
                sum += each[4]
                count += 1
        avr = sum / count
        avr_times[i] = avr
    return avr_times
```

```python
def avr_is_holiday(data):
    sum_is = count = sum_not = 0
    for i in range(len(data)):
        day = datetime.date(data[i][0], data[i][1], data[i][2])
        if (is_holiday(day)): 
            sum_is += data[i][4]
            count += 1
        else: 
            sum_not += data[i][4]
    return [sum_is / count, sum_not / (len(data) - count)]
```

```python
def is_weekend(day):
    if (day.weekday() == 4 or day.weekday() == 5):
        return True
    else: return False
```

```python
def avr_is_weekend(data):
    sum_is = count = sum_not = 0
    for i in range(len(data)):
        day = datetime.date(data[i][0], data[i][1], data[i][2])
        if (is_weekend(day)): 
            sum_is += data[i][4]
            count += 1
        else: 
            sum_not += data[i][4]
    return [sum_is / count, sum_not / (len(data) - count)]
```
